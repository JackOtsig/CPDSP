```python
import pandas as pd
import yfinance as yf
```

```python
tesla = yf.Ticker("TSLA")
tesla_share_price = tesla.history(period="max")
tesla_share_price.head()
```

```python
import bs4, requests, re, json
```

```python
tsla_revenue_data = pd.DataFrame(columns=["Date", "Revenue"])
with requests.get("https://companiesmarketcap.com/tesla/revenue/#:~:text=According%20to%20Tesla%27s%20latest%20financial,sale%20of%20goods%20or%20services.") as r:
    soup = bs4.BeautifulSoup(r.text, "html5lib")
data = str(soup.find_all("script")[6])
re1 = re.search('= ', data).span()
re2 = re.search(';', data).span()
real_data = data[re1[1]:re2[0]]
json_data = json.loads(real_data)
print(json_data)
for key in list(json_data.keys()):
    date = key
    revenue = json_data[key]
    tsla_revenue_data = tsla_revenue_data.append({"Date": date, "Revenue": revenue}, ignore_index=True)
tsla_revenue_data.tail()
```

```python
gme = yf.Ticker("GME")
gme_share_price = gme.history(period="max")
gme_share_price.head()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
fig = go.Figure(data=[go.Candlestick(x=list(tesla_share_price.index),
                                    open=tesla_share_price["Open"],
                                    high=tesla_share_price["High"],
                                    low=tesla_share_price["Low"],
                                    close=tesla_share_price["Close"])])
fig.update_layout(xaxis_rangeslider_visible=False)
fig.update_xaxes(title_text="TSLA Historical Data")
fig.show()
```
